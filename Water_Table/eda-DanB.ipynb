```python
import pandas as pd
from bokeh.io import push_notebook, output_notebook
from bokeh.plotting import figure, show
output_notebook()
```

```python
train_data = pd.read_csv('./tanzania-X-train.csv', header=0)
train_target = pd.read_csv('./tanzania-y-train.csv', header=0)
```

```python
train_data.info()
```

```python
train_data.describe
```

```python
train_data['status_group'] = train_target.status_group.astype('category')
```

```python
from bokeh.plotting import gmap
from bokeh.layouts import column
from bokeh.models import GMapOptions
from bokeh.palettes import brewer
from bokeh.models import CategoricalColorMapper
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral5
from bokeh.transform import linear_cmap
```

```python
mapoptions = GMapOptions(lat=train_data.latitude.median(), lng=train_data.longitude.median(), zoom=6)
```

```python
#colors = brewer["Spectral"][len(train_data.status_group.unique())]
#colormap = {x: colors[i] for i, x in enumerate(train_data.status_group.unique())}
#colors = [colormap[x] for x in train_data.status_group]
#categories = list(train_data.status_group.unique())
#color_mapper = CategoricalColorMapper(factors=categories)
#cmap = factor_cmap('status_group', palette=Spectral5, factors=sorted(train_data.status_group.unique()), end=1)
p1 = gmap('AIzaSyDdmZU-YmmPrBVhMarTxPBbEv7N_vdrXjY',mapoptions)
p1.circle(x='longitude', y='latitude', source=train_data[train_data['status_group'] == 'functional'],
         color='green')
p2 = gmap('AIzaSyDdmZU-YmmPrBVhMarTxPBbEv7N_vdrXjY',mapoptions)
p2.circle(x='longitude', y='latitude', source=train_data[train_data['status_group'] == 'functional needs repair'],
         color='yellow')
p3 = gmap('AIzaSyDdmZU-YmmPrBVhMarTxPBbEv7N_vdrXjY',mapoptions)
p3.circle(x='longitude', y='latitude', source=train_data[train_data['status_group'] == 'non functional'],
         color='red')
show(column(p1,p2,p3))
```

```python
p3 = gmap('AIzaSyDdmZU-YmmPrBVhMarTxPBbEv7N_vdrXjY',mapoptions)
p3.circle(x='longitude', y='latitude', source=train_data[train_data['construction_year'] != 0],
         color="green")
p3.circle(x='longitude', y='latitude', source=train_data[train_data['construction_year'] == 0],
         color="red")
show(p3)
```

```python
p3 = gmap('AIzaSyDdmZU-YmmPrBVhMarTxPBbEv7N_vdrXjY',mapoptions)
p3.circle(x='longitude', y='latitude', source=train_data[train_data['construction_year'] != 0],
         color=linear_cmap('construction_year', 'Plasma256', 1970, 2010))
show(p3)
```
